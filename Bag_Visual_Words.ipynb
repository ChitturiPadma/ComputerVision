```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
```

```python
imlist = {}
```

```python
path = '/home/padmach/Capabilities/ImageAnalytics/BagVisualWords/images/train/'
```

```python
count = 0
train_labels = np.array([])
name_dict = {}
descriptor_list = []
images = None
label_count = 0
sift_object = cv2.xfeatures2d.SIFT_create()
```

```python
from glob import glob
```

```python
for each in glob(path+'*'):
    word = each.split("/")[-1]
    #print word #displays sub-folders - dollar_bill, accordion, motorbike, Soccer_Ball
    imlist[word]=[]
    for imagefile in glob(path+word+'/*'):
        #print imagefile # displays image_0008.jpg, image_0013.jpg
        im = cv2.imread(imagefile,0)
        imlist[word].append(im)
        count +=1
print count
```

```python
for word, image in imlist.iteritems():
    name_dict[str(label_count)] =word
    for im in image:
        train_labels = np.append(train_labels, label_count)
        kp, des = sift_object.detectAndCompute(im, None)
        print(des.shape)
        descriptor_list.append(des)
    label_count +=1
```
